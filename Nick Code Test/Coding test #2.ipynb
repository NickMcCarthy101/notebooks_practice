```python
from collections import Counter 
def main():
    #Use open() for opening a file
    #Always use 'with' statement as it'll automatically close the file for you 
    with open(filename,'r') as f: #i.e having this as always equalling f (in this loop) 
        #text = f.read()
        #f.write(text.upper())
        words = [word for line in f for word in line.split()] #this splits the lines up and makes them into a list/flattens them
        
#         words = []
#         for line in f:
#             for word in line.split():      The code on the left is essentially the same as the code above
#                 words.append(word)
        
        #now use collections.Counter
        c = Counter(words)
        for word, count in c.most_common():
           print ("The word", word,"appears", count,"times in this document." )
main() #executes the function 

```

```python
from collections import Counter 
def main2():
    #Use open() for opening a file
    #Always use 'with' statement as it'll automatically close the file for you 
    with open(r'C:\\Users\\nick.mccarthy\\Desktop\\Nick Code Test\\doc2.txt') as f: 
        #create a list of all words fetched from the file using a list comprehension 
        words = [word for line in f for word in line.split()]
        #now use collections.Counter
        c = Counter(words)
        for word, count in c.most_common():
            print (word, count)
main2()   
```
