```python
from collections import Counter
def main():
    #Use open() for opening a file 
    #Always use 'with' statement as it'll automatically close the file for you 
    with open(filename,'r') as f: #i.e having this as always equalling f (in this loop)
        #text = f.read()
        #f.write(text.upper())
        words = [word for line in f for word in line.split()] #this splits the lines up and makes them into a list/flattens them 
        
        c= Counter(words)
        for word, count in c.most_common():
            print ("The word",word, "appears",count, "times in the document")
main() #Executes the function 
```
