```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.plot([1,2,3,4]) #seems to plot onto the y axis straight away
plt.show() #you can replace this with %matplotlibinline - just differnet info above the plot. 
```

```python
plt.plot([0.1,0.2,0.3,0.4], [1,2,3,4])
plt.show()
```

```python
#Can have multiple curves by putting as many x-y list pairs as needed
plt.plot([0.1,0.2,0.3,0.4], [1,2,3,4],
         [0.1,0.2,0.3,0.4], [1,4,9,16],[0.1,2],[0,2]) #doesn't matter if they're on the same line. Python knows to pair them up 
#immediately 
```

```python
plt.plot([0.1, 0.2, 0.3, 0.4], [1, 2, 3, 4],'rx', label ='first plot') #label references to the legend that is inputted
plt.plot([0.1, 0.2, 0.3, 0.4], [1, 4, 9, 16],'b-.', label = 'second plot' ) #the colour and style of the lines need to be 
plt.xlabel("Time (s)")                                                      #before the label 
plt.ylabel("Scale (Bananas)")
plt.legend()
plt.xlim(0,1)
plt.ylim(-5,20)
#see pyplot.plot for the list of colors and shapes
```
